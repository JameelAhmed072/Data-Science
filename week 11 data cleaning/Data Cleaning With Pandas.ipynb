## Working with Header

```python
df.columns
```

```python
# Removing () from the columns name which is best practice for data cleaning  
df.columns = df.columns.str.replace('(','').str.replace(')','')
```

```python
df.columns
```

```python
# Removing spaces from the begning and from the end of each columns names 
df.columns=df.columns.str.strip(" ")
```

```python
df.columns
```

```python
df.rename({'RAM':'RAM_GB'},axis=1,inplace=True)
```

```python
df.columns
```

```python
df.rename({'Screen Size':"Screen_Size_Inches"},axis = 1,inplace = True)
```

```python
df.columns
```

```python
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

## Data Cleaning

```python
# Removing GB from RAM_GB for better cleaning
```

```python
df.columns
```

```python
df['RAM_GB'] = df['RAM_GB'].str.replace('GB','') 
```

```python
df['RAM_GB']
```

```python
# Now let's comvert RAM_GB into integer
```

```python
df['RAM_GB'] = df['RAM_GB'].astype(int)
```

```python
df['RAM_GB']
```

```python
df['Screen_Size_Inches']=df['Screen_Size_Inches'].str.replace('"','')
```

```python
# Let's convert it's data type into float 
df['Screen_Size_Inches'] = df['Screen_Size_Inches'].astype(float)
```

```python
df['Screen_Size_Inches']
```
