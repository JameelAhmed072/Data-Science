```python
def average_daily_temp(year,months):
    tem = 0
    counter = 0
    average_temp = 0
    for i in dataset:
        date = i[0]
        if date.year == year and date.month == months:
            tem = tem + i[1]
            counter =counter + 1
    average_temp = tem/counter
    return average_temp
        
    
```
