```python
import pandas as pd
```

```python
df = pd.read_csv('Pokemon.csv')
```

```python
df
```

# Let's begin with the loc

```python
# loc is labeled based selecting method
```

```python
df.loc[4,'Name']
```

```python
df.loc[0:4]
```

```python
df.loc[0:4,['Name','Type 1']]
```

```python
df.loc[0:4,'Name':'Speed']
```

# Now let's come to iloc

```python
df.iloc[0:4,'Name':'Speed']
```

```python
# iloc is completely index based selecting method 
```

```python
df.iloc[1:5,0:4]
```

```python
df.iloc[[0,7]]   # this will only consider this as a row 0 and 7 and will print 0 to 7th column 
```

```python
df.iloc[:,[0,7]] # this is will now consider all rows due to : and will only bring 0 and 7th column
```
