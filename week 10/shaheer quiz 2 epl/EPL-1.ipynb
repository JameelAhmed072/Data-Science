```python
## Home Team Total Matches 

matches = {}
for i in data[1:]:
    home_name = i[2]
    away_name = i[3]
    season = i[10]
    if season == "2006-07":
        if home_name not in matches:
            matches[home_name] = 1
        else:
            matches[home_name] +=1
            
        
        
    
```

```python
# Home Team Total Matches
matches 
```

```python
## Away Team Total Matches and gather them with home team 
for i in data[1:]:
    home_name = i[2]
    away_name = i[3]
    season = i[10]
    if season == "2006-07":
        if away_name not in matches:
            matches[away_name] = 1
        else:
            matches[away_name] +=1
            
        
        
    
```

```python
matches # Total Matches of Both Teams 
```

```python
#sort Total Matches Dictionary
tot_matches = {}
so = sorted(matches)
for i in so:
    tot_matches[i] = matches[i]
```

```python
#sorted dictionary of Total Matches
tot_matches
```

```python
# convert Total Matches Dic into List 
tot_match = []
for key , val in tot_matches.items():
    tot_match.append([key , val])
    
tot_mat = tot_matches.values()    
```

```python
#compute Total wining matches that Home Team Win 

Tot_Wining = {}
for i in data[1:]:
    home_name = i[2]
    away_name = i[3]
    season = i[10]
    ftr = i[6]
    if season == "2006-07":
        if ftr == "H":
            if home_name not in Tot_Wining:
                Tot_Wining[home_name] = 1
            else:
                Tot_Wining[home_name] +=1
            
        
        
    
```

```python
## Compute Wining Matches of Away Team and gether it with Home Team
for i in data[1:]:
    home_name = i[2]
    away_name = i[3]
    season = i[10]
    ftr = i[6]
    if season == "2006-07":
        if ftr == "A":
            if away_name not in Tot_Wining:
                Tot_Wining[away_name] = 1
            else:
                Tot_Wining[away_name] +=1
            
        
        
    
```

```python
# Compute Total Goals Of Home Team Only
Tot_goals = {}

for i in data[1:]:
    home_name = i[2]
    away_name = i[3]
    season = i[10]
    FTHG = int(i[4])
    if season == "2006-07":
        if home_name not in Tot_goals:
            Tot_goals[home_name] = FTHG
        else:
            Tot_goals[home_name] +=FTHG
```

```python
# Compute goals against of away team
GA = {}
for i in data[1:]:
    home_name = i[2]
    away_name = i[3]
    FTHG = int(i[4])
    FTAG = int(i[5])
    season = i[10]
    ftr = i[6]
    if season == "2006-07":
        if away_name not in GA:
            GA[away_name] = FTHG
        else:
            GA[away_name] +=FTHG
```

```python
# Compute goals against of home team
GH = {}
for i in data[1:]:
    home_name = i[2]
    away_name = i[3]
    FTHG = int(i[4])
    FTAG = int(i[5])
    season = i[10]
    ftr = i[6]
    if season == "2006-07":
        if home_name not in GH:
            GH[home_name] = FTAG
        else:
            GH[home_name] +=FTAG
```
