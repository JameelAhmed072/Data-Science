```python
home_goals = {}
away_goals = {}
for i in dataset:
    season = i[10]
    FTHG = int(i[4])
    FTAG = int(i[5])
    if season == '2006-07':
        if i[2] not in home_goals:
            home_goals[i[2]] = FTHG
        else:
            home_goals[i[2]] += FTHG
            
        
        if i[3] not in away_goals:
            away_goals[i[3]] = FTAG
        else:
            away_goals[i[3]] += FTAG
        
            
            
```

```python
G_goals = {}
for i in home_goals:
    if i in away_goals:
        G_goals[i] = home_goals[i] + away_goals[i]
    else:
        pass
```

```python
GF = {}
a = sorted(G_goals.keys())
for i in a:
    GF[i] = G_goals[i]
```
