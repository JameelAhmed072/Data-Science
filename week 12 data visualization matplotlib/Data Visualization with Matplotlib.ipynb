```python
# Function by which we can check all the columns data yearly

def plot_cases(column_name):
    Name = df[column_name]
    
    plt.plot(df['Year'],Name)
    plt.title(column_name+" : Degree Graduates")
    plt.xlabel('Year')
    plt.ylabel('Percentage %')
    plt.show()
```

```python
plot_cases('Public Administration')
```

```python
plot_cases('Engineering')
```

```python
plot_cases('Math and Statistics')
```

```python
plot_cases('English')
```

```python
plot_cases('Health Professions')
```
