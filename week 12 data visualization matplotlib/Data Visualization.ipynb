```python
def plot_cases(country_name):
    country =  df.loc[df['Country'] == country_name]
    plt.plot(country['Date_reported'],country['Cumulative_cases'])
    plt.title(country_name + " Cumulative Cases")
    plt.xlabel(country_name + " Date Reported")
    plt.ylabel(country_name + " Cumulative Cases")
    plt.show()
    
```

```python
plot_cases('Pakistan')
```

```python
plot_cases('India')
```

```python
plot_cases('Bangladesh')
```
