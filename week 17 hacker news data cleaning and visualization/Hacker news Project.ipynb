# Now we will find average comments again but now we will do it wrt hours.

# Target is to know when to ask something on website (so we can get good response).

# By knowing the time we can engage more people and target more human traffic

```python
df['created_at']
```

```python
df['created_at'][0].hour
```

```python
df['created_at'].dt.hour
```

```python
df['created_at'].dt.strftime('%I %p')
```

```python
df['hour'] = df['created_at'].dt.strftime('%I %p')
```

```python
df['hour']
```

```python
askhn = df[df['Title Category'] == 'ASK HN']
showhn = df[df['Title Category'] == 'SHOW HN'] 
```

```python
askhn
```

```python
showhn
```

```python
askhn = askhn.pivot_table(index = 'hour', values = 'num_comments', aggfunc = 'mean')
```

```python
# askhn
```

```python
askhn.sort_values(by = 'num_comments', ascending = False)[:5]
```

```python
# askhn
```

```python
showhn = showhn.pivot_table(index = 'hour', values = 'num_comments', aggfunc = 'mean')
```

```python
showhn.sort_values(by = 'num_comments', ascending = False)[:5]
```

```python
# showhn
```

# Now let's show this graphically 

```python
sns.reset_defaults()
sns.lineplot(data = askhn, x = askhn.index, y = 'num_comments')
plt.tick_params(axis = 'x', rotation = 90)
plt.show()
```
