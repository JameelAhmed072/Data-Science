```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing = pd.read_csv("Housing.csv")
```

```python
housing.head()
```

```python
housing.info()
```

# Living Area

```python
sns.set_theme()

sns.relplot(data = housing , x = "Gr Liv Area" , y = "SalePrice")

plt.show()
```

# Overall Quality

```python
housing["Overall Qual"]
```

```python
housing["Overall Qual"].unique()
```

```python
sns.set_theme()

sns.relplot(data = housing , x = "Gr Liv Area" , y = "SalePrice" , hue = "Overall Qual")

plt.show()
```

```python
sns.set_theme()

sns.relplot(data = housing , x = "Gr Liv Area" , y = "SalePrice" , hue = "Overall Qual",\
           palette = "RdYlBu")

plt.show()
```

```python
housing.corr()["SalePrice"].sort_values(ascending = False)
```

# Garage Area

```python
sns.set_theme()

g = sns.relplot(data = housing , x = "Gr Liv Area" , y = "SalePrice" , hue = "Overall Qual",\
           palette = "RdYlBu" , size = "Garage Area" , sizes = (1,200))

g.fig.set_figheight(8)
g.fig.set_figwidth(15)


plt.show()
```

# Rooms

```python
sns.set_theme()

g = sns.relplot(data = housing , x = "Gr Liv Area" , y = "SalePrice" , hue = "Overall Qual",\
           palette = "RdYlBu" , style = "Rooms" , size = "Garage Area" , sizes = (1,200))

g.fig.set_figheight(8)
g.fig.set_figwidth(15)


plt.show()
```

# Old Construction vs New Construction

```python
sns.set_theme()

g = sns.relplot(data = housing , x = "Gr Liv Area" , y = "SalePrice" , hue = "Overall Qual",\
           palette = "RdYlBu" , col = "Year" , style = "Rooms" , size = "Garage Area" , sizes = (1,200))

g.fig.set_figheight(8)
g.fig.set_figwidth(15)


plt.show()
```
