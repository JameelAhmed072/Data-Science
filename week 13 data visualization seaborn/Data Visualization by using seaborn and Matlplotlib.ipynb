```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('housing.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

# Living Area


```python
df['Gr Liv Area']
```

```python
df['SalePrice']
```

```python
sns.set_theme()
g = sns.relplot(data = df ,  x = 'Gr Liv Area', y = 'SalePrice')

g.fig.set_figheight(8)
g.fig.set_figwidth(15)
plt.show()
```

# Overall Quality

```python
df['Overall Qual'].unique()
```

```python
df['Overall Qual']
```

```python
sns.set_theme()
g = sns.relplot(data = df ,  x = 'Gr Liv Area', y = 'SalePrice',hue =  'Overall Qual')

g.fig.set_figheight(8)
g.fig.set_figwidth(15)
plt.show()
```

```python
sns.set_theme()
g = sns.relplot(data = df ,  x = 'Gr Liv Area', y = 'SalePrice',hue =  'Overall Qual',palette = 'RdYlBu')

g.fig.set_figheight(8)
g.fig.set_figwidth(17)
plt.show()
```

```python
df.corr()['SalePrice'].sort_values(ascending = False)
```

# Garage Area

```python
df.columns
```

```python
sns.set_theme()
g = sns.relplot(data = df ,  x = 'Gr Liv Area', y = 'SalePrice',hue =  'Overall Qual',palette = 'RdYlBu',size = 'Garage Area'\
               ,sizes = (1,300))

g.fig.set_figheight(8)
g.fig.set_figwidth(15)
plt.show()
```

```python
sns.set_theme()
g = sns.relplot(data = df ,  x = 'Gr Liv Area', y = 'SalePrice',hue =  'Overall Qual',palette = 'RdYlBu',size = 'Garage Area'\
               ,sizes = (1,300),style = 'Rooms')

g.fig.set_figheight(8)
g.fig.set_figwidth(15)
plt.show()
```

# Old Construction or new Construction

```python
sns.set_theme()
g = sns.relplot(data = df ,  x = 'Gr Liv Area', y = 'SalePrice',hue =  'Overall Qual',palette = 'RdYlBu',size = 'Garage Area'\
               ,col = 'Year',sizes = (1,300),style = 'Rooms')

g.fig.set_figheight(8)
g.fig.set_figwidth(15)
plt.show()
```
