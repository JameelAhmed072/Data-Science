```python
# First of all convert the Slowness in traffic which is object to change it into float
```

```python
df['Slowness in traffic (%)'] = df['Slowness in traffic (%)'].str.replace(',','') 
```

```python
df['Slowness in traffic (%)'] = df['Slowness in traffic (%)'].astype(float) 
```

```python
df.dtypes
```
