```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('housing.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df.info()
```

```python
sns.set_theme()
g = sns.relplot(data = df,x = 'Gr Liv Area',y = 'SalePrice')
g.fig.set_figheight(8)
g.fig.set_figwidth(15)

plt.show()
```

```python
df['Overall Qual'].unique()
```

```python
sns.set_theme()
g = sns.relplot(data = df,x = 'Gr Liv Area',y = 'SalePrice',hue = 'Overall Qual')
g.fig.set_figheight(8)
g.fig.set_figwidth(15)

plt.show()
```

```python
sns.set_theme()
g = sns.relplot(data = df,x = 'Gr Liv Area',y = 'SalePrice',hue = 'Overall Qual',palette='')
g.fig.set_figheight(8)
g.fig.set_figwidth(15)

plt.show()
```

```python
sns.set_theme()
g = sns.relplot(data = df,x = 'Gr Liv Area',y = 'SalePrice',hue = 'Overall Qual',palette='crest')
g.fig.set_figheight(8)
g.fig.set_figwidth(15)

plt.show()
```

```python
df.corr()['SalePrice'].sort_values(ascending=False)
```

```python
df['Garage Area'].unique()
```

```python
sns.set_theme()
g = sns.relplot(data = df,x = 'Gr Liv Area',y = 'SalePrice',hue = 'Overall Qual',palette='crest',size = 'Garage Area',
               sizes = (1,300))
g.fig.set_figheight(8)
g.fig.set_figwidth(15)

plt.show()
```

```python
df['Rooms'].unique()
```

```python
sns.set_theme()
g = sns.relplot(data = df,x = 'Gr Liv Area',y = 'SalePrice',hue = 'Overall Qual',palette='crest',size = 'Garage Area',
               sizes = (1,300),style='Rooms')
g.fig.set_figheight(8)
g.fig.set_figwidth(15)
plt.show()
```

```python
sns.set_theme()
g = sns.relplot(data = df,x = 'Gr Liv Area',y = 'SalePrice',hue = 'Overall Qual',palette='crest',size = 'Garage Area',
               sizes = (1,300),style='Rooms',col='Year')
g.fig.set_figheight(8)
g.fig.set_figwidth(15)
plt.show()
```
