```python
free_apps = []
paid_apps = []

for i in dataset[1:]:
    
    name = i[1]
    price = float(i[4])
    
    #print(name , price)
    
    if price == 0:
        free_apps.append(name)
        
    else:
        paid_apps.append(name)
```

### Rating of free apps and paid apps

```python
free_apps_rating = []
paid_apps_rating = []

for i in dataset[1:]:
    
    price = float(i[4])
    rating = float(i[7])
    
    if price == 0: # app is free
        free_apps_rating.append(rating)
        
    else:
        paid_apps_rating.append(rating)
```

```python
sum(free_apps_rating)/len(free_apps_rating)
```

```python
sum(paid_apps_rating)/len(paid_apps_rating)
```

### Get Me Games That are Free Of Cost

```python
for i in dataset[1:]:
    
    app_name = i[1]
    price = float(i[4])
    genre = i[11]
    
    if price == 0 and genre == "Games":
        print(app_name , "====" , price)
```
