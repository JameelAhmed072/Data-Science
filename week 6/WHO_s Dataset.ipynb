```python
def readfile(filename):
    from csv import reader
    return list(reader(open(filename,encoding = 'utf-8')))
```

```python
readfile('WHO.csv')
```

```python
dataset = readfile('WHO.csv')
```

# New Deaths Country wise 

```python
dataset[0].index('Country')
```

```python
dataset[0].index('New_deaths')
```

```python
newdeaths_countries = {}
for i in dataset[1:]:
    
    country = i[1]
    
    new_deaths = int(i[4])
    
    if country not in newdeaths_countries:
        
        newdeaths_countries[country] = 0
        
    else:
        newdeaths_countries[country] += new_deaths
    
```
