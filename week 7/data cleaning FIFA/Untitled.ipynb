```python
def readfile(filename):
    from csv import reader
    return list(reader(open(filename,encoding = 'utf-8')))
```

```python
dataset = readfile('fifa21_raw_data.csv')
```

```python
header = dataset[0]
```

```python
dataset = dataset[1:]
```

```python
portugal_players = []
for i in dataset:
    nat = i[3]
    if nat == 'Portugal':
        portugal_players.append(i[1])
```

```python
portugal_players
```

```python
freq_players = {}
for i in dataset:
    nat = i[3]
    if nat not in freq_players:
        freq_players[nat] = 1
    else:
        freq_players[nat] += 1 
```
