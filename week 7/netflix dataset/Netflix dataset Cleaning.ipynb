# Removing bad characters from the release_year column

```python
bad_char = ["(" , ")" , "c" , "C" , "." , " " , "S" , "s"]
```

```python
for i in bad_char:
    print(i)
```

```python
def clean_bad_char(n):
    for i in bad_char:
        
        n = n.replace(i,"")
        
    return n    
```

# Splitting the last Column

```python
for i in dataset:
    print(i[5])
```

```python
def splitting(n):
    lisst = i[n]
    if ',' in lisst:
        total = lisst.split(',')
        first_index = total[0]
        lisst = first_index
        i[n] = lisst
        return n
```

```python
for i in dataset:
    splitting(5)
```
