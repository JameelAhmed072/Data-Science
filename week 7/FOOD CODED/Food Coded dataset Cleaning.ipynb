# Breakfast Column

```python
def conversion_breakfast(n):
    breakfast = i[n]
    if breakfast == '1':
        breakfast = breakfast.replace('1','Once')
        i[n] = breakfast
    elif breakfast == '2':
        breakfast = breakfast.replace('2','Twice')
        i[n] = breakfast
```

# Calories Column

```python
header
```

```python
header.index('calories_chicken')
```

```python
def calories_chicken(n):
    cal = i[n]
    if cal == '265':
        cal = cal.replace('265','Low')
        i[n] = cal
    elif cal == '430':
        cal = cal.replace('430','medium calories')
        i[n] = cal
    elif cal == '610':
        cal = cal.replace('610','High')
        i[n] = cal
    elif cal == '720':
        cal = cal.replace('720','very High')
        i[n] = cal
            
```

```python
for i in dataset:
    calories_chicken(3)
```
