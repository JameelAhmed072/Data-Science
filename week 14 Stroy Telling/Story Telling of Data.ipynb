# Removing Labels

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,1,figsize=(6,8))


axes = [ax1,ax2,ax3,ax4]
for ax in axes:
    ax.plot(df['Month'],df['New_deaths'])
    
    ax.set_yticklabels([])
    ax.set_xticklabels([])    
```

# Removing Params/sticks

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,1,figsize=(6,8))


axes = [ax1,ax2,ax3,ax4]
for ax in axes:
    ax.plot(df['Month'],df['New_deaths'])
    
    ax.set_yticks([])
    ax.set_xticks([])    
    
    ax.tick_params(left = 0, bottom = 0)
```

# Removing spines/Borders

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,1,figsize=(6,8))


axes = [ax1,ax2,ax3,ax4]
for ax in axes:
    ax.plot(df['Month'],df['New_deaths'])
    
    ax.set_yticks([])
    ax.set_xticks([])    
    
    ax.tick_params(left = 0, bottom = 0)
    
    
    locations = ['left','bottom','right','top']
    for i in locations:
        ax.spines[i].set_visible(False)
```

# Adding Color

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,1,figsize=(6,8))


axes = [ax1,ax2,ax3,ax4]
for ax in axes:
    ax.plot(df['Month'],df['New_deaths'],color = '#b03030', alpha = 0.4)  # Alpha defines the opacity
    
    ax.set_yticks([])
    ax.set_xticks([])    
    
    ax.tick_params(left = 0, bottom = 0)
    
    
    locations = ['left','bottom','right','top']
    for i in locations:
        ax.spines[i].set_visible(False)
```
