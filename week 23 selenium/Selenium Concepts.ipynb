# part 1 Set up

```python
# relevant packages & modules
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
```

```python
website = "https://www.daraz.pk/catalog/?q=watch&_keyori=ss&from=input&spm=a2a0e.home.search.go.35e3493775HQzW"
```

```python
# initilize chrome
driver = webdriver.Chrome("C:\webdrivers\chromedriver.exe")
# open website
driver.get(website)
# maximize windows
driver.maximize_window()
```

# part 2 locators

### xPath-Locator

### <span class="currency--GVKjl">Rs. 689</span>

### watch prices

```python
price = driver.find_element('xpath','//span[@class="currency--GVKjl"]')         # inplace of xpath we can also use BY.CLASS_NAME
```

```python
price.text
```

```python
prices = driver.find_elements('xpath','//span[@class="currency--GVKjl"]')
len(prices)
```

```python
for i in prices:
    print(i.text)
```

## class Name locator

### watch names

```python
names = driver.find_elements(By.CLASS_NAME,'title--wFj93')
len(names)
```

```python
for i in names:
    print(i.text)
    print()
```

## ID -Locator


### Search Bar

```python
search = driver.find_element('id','q')
```

```python
search
```

```python
#click
search.click()
time.sleep(2)

# selecting 
search.send_keys(Keys.CONTROL + 'a')

#erasing 
search.send_keys(Keys.DELETE)

# searching now
search.send_keys('books')
```

## Name Locator
#### Entering on search bar

```python
search = driver.find_element('name','q')


#click
search.click()
time.sleep(2)

#entering 
search.send_keys(Keys.ENTER)
```

#### Going Back

```python
driver.back()
```
