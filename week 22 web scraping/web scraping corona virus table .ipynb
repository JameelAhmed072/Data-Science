```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_site = "https://www.worldometers.info/coronavirus/"
```

```python
response = requests.get(base_site)
```

```python
response
```

```python
html = response.content
```

```python
# html
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
# soup
```

```python
table = soup.find('table',id = 'main_table_countries_today')
```

```python
# table
```

```python
headers = []
for i in table.find_all('th'):
    title = i.text
    headers.append(title)
```

```python
headers
```

```python
headers[13]
```

```python
headers[13] = 'Tests/1M pop'
```

```python
mydata = pd.DataFrame(columns = headers)
```

```python
for j in table.find_all('tr')[1:]:
    row_data = j.find_all('td')
    row = []
    
    for i in row_data:
        row.append(i.text)
        
    length = len(mydata)
    mydata.loc[length] = row
    
```

```python
# Drop and clearing unnecessary rows
mydata.drop(mydata.index[0:7], inplace=True)
mydata.drop(mydata.index[222:229], inplace=True)
mydata.reset_index(inplace=True, drop=True)

# Drop “#” column
mydata.drop('#', inplace=True, axis=1)
```

```python
# Export to csv
mydata.to_csv('Covid19_data.csv', index=False)

# Try to read csv
mydata2 = pd.read_csv('Covid19_data.csv')
```
