# Now let's apply the category on the original dataframe


```python
def category(e,a):
    if e > a:
        return 'High'
    else:
        return 'Low'
```

```python
df['Economy (GDP per Capita)'].apply(category,a=1)
```

```python
# Apply it by using the column name
```

```python
def category(element,x,col_name):
    if element > x:
        return f'High {col_name}'
    else:
        return f"Low {col_name}"
```

```python
df['Economy (GDP per Capita)'].apply(category , x = 1, col_name = "Economy")
```

# Pivot table

```python
df.pivot_table(index = 'Region', values = 'Happiness Score',aggfunc = 'mean')
```

```python
df.pivot_table(index = 'Region', values = 'Happiness Score',aggfunc = 'mean').sort_values(by = 'Happiness Score',ascending = False)
```

```python
df.pivot_table(index = 'Region', values = ['Happiness Score','Happiness Rank'],aggfunc = ['mean','std','min','max','count']) 
```
